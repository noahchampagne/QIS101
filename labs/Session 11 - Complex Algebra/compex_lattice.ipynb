```python
"""complex_lattice.ipynb"""
# Cell 1

from __future__ import annotations

import typing

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.collections import LineCollection

if typing.TYPE_CHECKING:
    from matplotlib.axes import Axes
    from numpy.typing import NDArray

%matplotlib widget


def f(x: complex) -> complex:
    """Given Function"""
    return x**2 + complex(0, 1) * x + 1


def plot(ax: Axes) -> None:
    """Plots the real and imaginary parts of the given function"""
    # Uses list comprehension to create all the input and output values
    x_vals: NDArray[np.complex_] = np.array(
        [complex(i, j) for i in np.linspace(-4, 4, 20) for j in np.linspace(0, 2, 20)])
    results: NDArray[np.complex_] = f(x_vals)

    # Plots the graph and sets the labels
    plt.scatter(results.real, results.imag)
    ax.set_title(r"$z^2 + iz + 1$")
    ax.set_xlabel("Imaginary Part")
    ax.set_ylabel("Real Part")
    ax.set_aspect("equal")


def main() -> None:
    plt.figure(" ")
    plot(plt.axes())
    plt.show()


main()
```

```python
# Cell 2

from __future__ import annotations

import typing

import numpy as np

if typing.TYPE_CHECKING:
    from numpy.typing import NDArray


def f(x: complex) -> complex:
    """Given Function"""
    return x**2 + complex(0, 1) * x + 1


def solve() -> None:
    """Plots the real and imaginary parts of the given function"""
    # Uses list comprehension to create all the input and output values
    x_vals: NDArray[np.complex_] = np.array(
        [complex(i, j) for i in np.linspace(-10, 10, 21) for j in np.linspace(0, 10, 11)])
    results: NDArray[np.complex_] = f(x_vals)

    # Count the number of guassian integers int the desired range
    count = 0
    for num in results:
        real: float = num.real
        imag: float = num.imag
        if -10 < real < 10 and -10 < imag < 10:
            count += 1
    print(count)


def main() -> None:
    solve()


main()
```
